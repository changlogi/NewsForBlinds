```python
# 크롤링 모듈 로드

from bs4 import BeautifulSoup
import requests
import re
```

```python
#링크가 주어지면, 기사 제목과 본문을 크롤링하는 코드
def get_news_by_url(url):

    headers={
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
        'accept-encoding': 'gzip, deflate, br',
        'accept-language': 'ko,en;q=0.9,en-US;q=0.8',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36 Edg/105.0.1343.53'
    }
    res=requests.get(url,headers=headers)
    bs=BeautifulSoup(res.content,'html.parser')

    #제목 찾아오기 (html 분석하여 기사 제목 담고 있는 html tag  지정)
    title=bs.find('div','article_info').find('h3').string
    title=re.sub('\t|\r|\n|','',title)

    #본문 찾아오기 (html 분석하여 기사 본문 담고 있는 html tag  지정)
    content=bs.find('div','articleCont').text
    content=re.sub('\xa0|\t|\r|\n|','',content)

    return title+content
```

```python
from konlpy import init_jvm
from konlpy.tag import Kkma
from konlpy.tag import Okt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import normalize
import numpy as np
```

```python
# 꼬꼬마 분석기
kkma = Kkma()

def text2sentences(text):
    sentences = kkma.sentences(text)
    for idx in range(0, len(sentences)):
        if len(sentences[idx]) <= 10:
            sentences[idx-1] += (' ' + sentences[idx])
            sentences[idx] = ''
    return sentences
```

```python
# 기사본문 가져오기
sentences = text2sentences(news_content[0])
```

```python
twitter = Okt()

# 사용하지 않을 단어 제거
stopwords = ['한국경제','매일경제','머니투데이','연합뉴스','데일리','동아일보','중앙일보',
             '기자','아','휴','아이구','아이쿠','어','나','우리','저희','따라','의해','을',
             '를','에','의','가','지금','과거','뉴스','현재','망치','매우','실제','전후',
             '안팎','제외','불과','매일']

def get_nouns(sentences):
    nouns = []
    for sentence in sentences:
        if sentence != '':
            nouns.append(' '.join([noun for noun in twitter.nouns(str(sentence)) if noun not in stopwords and len(noun) > 1]))
    return nouns
```

```python
nouns = get_nouns(sentences)
```

```python
tfidf = TfidfVectorizer()
cnt_vec = CountVectorizer()
graph_sentence = []

def build_sentences_graph(sentence):
    tfidf_mat = tfidf.fit_transform(sentence).toarray()
    graph_sentence = np.dot(tfidf_mat, tfidf_mat.T)
    return graph_sentence

sentences_graph = build_sentences_graph(nouns)
```

```python
print(sentences_graph)
```

```python
def build_words_graph(sentence):
    cnt_vec_mat = normalize(cnt_vec.fit_transform(sentence).toarray().astype(float), axis=0)
    vocab = cnt_vec.vocabulary_
    return np.dot(cnt_vec_mat.T, cnt_vec_mat), {vocab[word] : word for word in vocab}

words_graph, idx2word = build_words_graph(nouns)
```

```python
def get_ranks(graph, d=0.85):
    
    A = graph
    matrix_size = A.shape[0]
    for id in range(matrix_size):
        A[id, id] = 0
        link_sum = np.sum(A[:,id])
        if link_sum != 0:
            A[:, id] /= link_sum
        A[:, id] *= -d
        A[id, id] = 1
    
    B = (1-d) * np.ones((matrix_size, 1))
    ranks = np.linalg.solve(A, B)

    return{idx: r[0] for idx, r in enumerate(ranks)}
```

```python
sentences_rank_idx = get_ranks(sentences_graph)
sentences_rank_idx
```

```python
sorted_sentences_rank_idx = sorted(sentences_rank_idx, key=lambda k: sentences_rank_idx[k], reverse=True)
```

```python
words_rank_idx = get_ranks(words_graph)
sorted_words_rank_idx = sorted(words_rank_idx, key=lambda k: words_rank_idx[k], reverse=True)
```

```python
def summarize(sentences_num=5):
    summary = []
    index = []
    sentence = ''
    for idx in sorted_sentences_rank_idx[:sentences_num]:
        index.append(idx)
        
    index.sort()
    
    for idx in index:
        summary.append(sentences[idx])
    
    for text in summary:
        print(text)
        print("\n")
        sentence += text
    return sentence
```
